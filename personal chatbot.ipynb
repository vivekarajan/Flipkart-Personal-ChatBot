```python
def extract_date():
    df['crawl_timestamp'] = pd.to_datetime(df['crawl_timestamp'], format='%Y-%m-%d %H:%M:%S %z')
    df['Date'] = df['crawl_timestamp'].dt.date
    df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
    return df.dtypes
```

# using Voice Recognition

```python
import pyttsx3
import speech_recognition as sr
import datetime
import pandas as pd
engine = pyttsx3.init()

def speak(text):
    engine.say(text)
    engine.runAndWait()

def greeting():
    hour = datetime.datetime.now().hour
    if 6 <= hour < 12:
        speak("Good morning")
    elif 12 <= hour < 18:
        speak("Good afternoon")
    elif 18 <= hour < 24:
        speak("Good evening")
    else:
        speak("Good night")

def wish():
    speak("Welcome back sir")
    greeting()
    speak("Jarvis at your service, please tell me how can I help you?")

def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        speak("Listening...")
        r.pause_threshold = 1
        audio = r.listen(source)

    try:
        query = r.recognize_google(audio, language='en-in')
        print(f"User said: {query}\n")
    except Exception as e:
        print(e)
        speak("Sorry, I didn't catch that. Could you please say that again?")
        return "None"
    return query

class Details:
    #top 5 brands by count
    @staticmethod
    def Brand_Count():
        brand_counts = df['brand'].value_counts()
        top_5_brands = brand_counts.head(5)
        print(top_5_brands)
        speak(f"The top 5 brands are: {', '.join(top_5_brands.index)}")

    #Average discount percentage
    @staticmethod
    def Avg_Dis():
        df['discount_percentage'] = ((df['retail_price'] - df['discounted_price']) / df['retail_price']) * 100
        average_discount_percentage = df['discount_percentage'].mean()
        print("Average Discount Percentage:", average_discount_percentage)
        speak(f"The average discount percentage is {average_discount_percentage:.2f} percent")
    
    #Number of ratings available
    @staticmethod
    def rating():
        filtered_df = df[df['product_rating'] != 'No rating available']
        count_of_ratings = filtered_df['product_rating'].count()
        print("Number of Ratings Available:", count_of_ratings)
        speak(f"There are {count_of_ratings} ratings available")
    
    #Top 3 most expensive items
    @staticmethod
    def Expensive():
        top_3_expensive = df.nlargest(3, 'retail_price')
        print(top_3_expensive)
        speak(f"The top 3 most expensive items are priced at {', '.join(map(str, top_3_expensive['retail_price'].values))}")

command_mapping = {
    'brand count': Details.Brand_Count,
    'average discount': Details.Avg_Dis,
    'ratings available': Details.rating,
    'expensive items': Details.Expensive
}

def main():
    wish()
    while True:
        query = takeCommand().lower()
        
        if 'brand count' in query:
            Details.Brand_Count()
        elif 'average discount' in query:
            Details.Avg_Dis()
        elif 'ratings available' in query:
            Details.rating()
        elif 'expensive items' in query:
            Details.Expensive()
        elif 'stop' in query:
            speak("Ok Goodbye, Sir.")
            break

if __name__ == "__main__":
    main()
```
